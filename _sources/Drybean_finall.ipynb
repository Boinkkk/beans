### PEMERIKSAAN MISSING VALUE

Langkah pertama untuk mengidentifikasi kualitas data salah satunya dengan pemeriksaan missing values. Pemeriksaan missing values sangat penting karena nilai yang hilang dapat mempengaruhi analisis dan model yang dibangun. Mengidentifikasi dan menangani missing values adalah langkah krusial dalam proses pembersihan data. Berikut adalah kode untuk memeriksa missing values dalam dataset:

```python
def check_missing_value(data:pd.DataFrame) -> pd.DataFrame or str:
    if isinstance(data, pd.DataFrame) is False:
        return print("Data yang di input bukan dataframe")
    else:
        is_missing = data.isnull().values.any()
        total_missing = data.isnull().sum().sum()
        df = {
            'Missing Value?' : [is_missing],
            'Total' : [total_missing]
        }
        return pd.DataFrame(df, index=['Result'])

check_missing_value(dataset)
```

### PEMERIKSAAN OUTLIER

Deteksi outlier adalah metode untuk mengidentifikasi titik-titik data yang dianggap sebagai outlier atau anomali. Outlier adalah nilai yang secara signifikan berbeda dari sebagian besar data lainnya dalam sebuah dataset. Mengidentifikasi outliers bermanfaat untuk meningkatkan akurasi analisis dan model prediktif, pada kali akan menggunakan euclidian distance

Pada Kasus dataset kali ini kita akan menghapus hanya 1 nilai saja dikarenakan anomaly score yang terlalu tinggi sehingga kemungkinan besar data itu mengalami error saat ektraksi fitur. kita tidak melakukan penghapus terhadap nilai yang "dianggap outlier" dikarenakan hal itu tidak diperlukan, data yang beranomaly score tinggi sebetulnya hanyalah variasi dari Dry Bean tersebut layaknya manusia yang membunyai berat badan berlebih dan ada juga yang kekurangan. pengapusan outlier juga membuat ruang lingkup belajar model lebih sedikit sehingga rawan mengalami overfitting

<img src="https://deepnote.com/api/project/e36549a5-e39b-4420-be70-e5f5c5b65c33/download-file?path=image-20240627-064011.png" width="" align="" />

```python
import matplotlib.pyplot as plt
from sklearn.neighbors import LocalOutlierFactor
import numpy as np

def make_plot_scatter(x:list, y:list, x_title:str, y_title:str, title:str, outlier:bool=False,outlier_value = None):
    if len(x) == len(y) is False:
        return print('Jumlah Data di X dan Y tidak sama!')
    else:
        plt.style.use('fivethirtyeight')
        plt.figure(figsize=(10,6))
        plt.title(title)
        plt.xlabel(x_title)
        plt.ylabel(y_title)
        if outlier:
            plt.scatter(x, y, c=(outlier_value == -1), cmap='coolwarm', edgecolors='k')
            return plt.show()
        else:
            plt.scatter(x, y)
            plt.legend()
            return plt.show()

def get_outlier(data:pd.DataFrame, n_neighbors:int = 20,contamination:float=0.040) -> pd.DataFrame:
    data = data.select_dtypes(include='number')
    lof = LocalOutlierFactor(n_neighbors=n_neighbors, contamination=contamination, metric='euclidean')
    outlier = lof.fit_predict(data)
    outlier = pd.DataFrame(outlier)
    return outlier

def get_anomaly_scores(data:pd.DataFrame,n_neighbors:int = 20,contamination:float=0.040):
    data = data.select_dtypes(include='number')
    lof = LocalOutlierFactor(n_neighbors=n_neighbors, contamination=contamination, metric='euclidean')
    lof.fit_predict(data)
    return pd.DataFrame(lof.negative_outlier_factor_)

dataset_anomaly_scores = get_anomaly_scores(dataset, 20, 0.00001)
outlier = get_outlier(dataset, 20, 0.00001)
y = dataset_anomaly_scores
X = np.arange(len(dataset))


make_plot_scatter(X, y, 'x', 'y', 'outlier', True, outlier)
```

## PENGHAPUS OUTLIER

Langkah pertama adalah menghapus outlier menggunakan metode Local Outlier Factor (LOF) dari dataset. Penghapusan outlier sangat penting untuk membersihkan data dari nilai yang tidak lazim atau anomali yang dapat mempengaruhi hasil analisis. Outlier diidentifikasi dengan kondisi di mana nilai 'LOF_Prediksi' tidak sama dengan -1. Dan Berikut adalah langkah-langkahnya:

1) memeriksa setiap baris dalam kolom 'LOF_Prediksi' dari DataFrame df.
2) Baris-baris yang memiliki nilai tidak sama dengan -1 dianggap sebagai outlier dan akan dihapus.
3) Pada data frame baru dibuat hanya dengan baris-baris yang tidak dianggap sebagai outlier
4) Setelah menghapus outlier, jumlah baris data yang tersisa adalah 13610 dari total awal 13611 baris.

Kode di bawah digunakan untuk menghapus baris-baris yang dianggap sebagai outlier dari sebuah dataset menggunakan prediksi dari metode LOF.

```python
def delete_outlier(data:pd.DataFrame, outlier_name:str):
    clean_data = data[data[outlier_name] != -1].reset_index(drop=True)
    clean_data = clean_data.drop(columns=[outlier_name])
    return clean_data

dataset['Outlier'] = outlier
print ('Jumlah baris sebelum penghapusan outlier: %d' % (len(dataset)))
display(dataset)
dataset_clean = delete_outlier(dataset, 'Outlier')
print('Jumlah baris setelah penghapusan outlier: %d' % (len(dataset_clean)))
display(dataset_clean)
```

### METODE STANDARISASI DATA

Normalisasi adalah proses mengubah skala nilai fitur dalam dataset sehingga semua nilai berada dalam rentang tertentu, biasanya antara 0 dan 1. Normalisasi membantu dalam memastikan bahwa semua fitur memiliki bobot yang sama ketika digunakan dalam algoritma machine learning. Ini sangat penting ketika fitur memiliki unit atau skala yang berbeda. Misalnya, jika satu fitur memiliki rentang nilai antara 1 hingga 1000 dan fitur lainnya antara 0 hingga 1, algoritma machine learning mungkin akan lebih mengutamakan fitur yang memiliki skala lebih besar. Dengan normalisasi, skala nilai fitur dapat disamakan, memungkinkan algoritma untuk bekerja lebih efektif dan seimbang.

### METODE NORMALISASI DATA

Standarisasi, juga dikenal sebagai z-score normalization, adalah proses mengubah nilai fitur dalam dataset sehingga memiliki mean (rata-rata) 0 dan standard deviation (simpangan baku) 1. Ini dilakukan dengan mengurangi mean dari setiap nilai fitur dan kemudian membaginya dengan standard deviation. Standarisasi sangat berguna ketika dataset memiliki distribusi normal (Gaussian). Dengan standarisasi, kita dapat memastikan bahwa fitur memiliki kontribusi yang seimbang dalam algoritma machine learning, terutama yang sensitif terhadap skala data, seperti regresi linear dan support vector machines (SVM).

```python
from sklearn.preprocessing import MinMaxScaler
scaler_minmax = MinMaxScaler()
scaler_minmax.fit(X)
X_minmax = scaler_minmax.transform(X)
X_minmax = pd.DataFrame(X_minmax, columns=X.columns)
display(X_minmax)
```

```python
from sklearn.naive_bayes import GaussianNB, CategoricalNB, BernoulliNB, MultinomialNB,ComplementNB
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
def get_accuracy(method, X_train, y_train, X_test, y_test):
    le = LabelEncoder()
    y_train = le.fit_transform(y_train)
    y_test = le.fit_transform(y_test)
    acc = []
    index =  []
    for clf in method:
        c = clf
        c.fit(X_train, y_train.ravel())
        y_pred = c.predict(X_test)
        accucary = accuracy_score(y_test, y_pred)
        acc.append(accucary)
        index.append(str(clf).replace('(', '').replace(')', ''))

    df = pd.DataFrame({
        'Model': index,
        'Accuracy': acc

    })
    return df

method = [
    GaussianNB(),
    BernoulliNB(),
    MultinomialNB(),
    ComplementNB()
]
minmax = get_accuracy(method, X_train_minmax, y_train_minmax, X_test_minmax, y_test_minmax)
display(minmax)
```

### TREE


#### Apa itu Decision Tree?
Decision tree adalah struktur mirip diagram alir yang digunakan untuk membuat keputusan atau prediksi. Struktur ini terdiri dari node yang mewakili keputusan atau pengujian pada atribut, cabang yang mewakili hasil dari keputusan ini, dan leaf nodes yang mewakili hasil akhir atau prediksi. Setiap node internal sesuai dengan pengujian pada atribut, setiap cabang sesuai dengan hasil pengujian, dan setiap leaf node sesuai dengan label kelas atau nilai kontinu.

#### Struktur Decision Tree
- Root Node: Mewakili seluruh dataset dan keputusan awal yang harus dibuat.
- Internal Nodes: Mewakili keputusan atau pengujian pada atribut. Setiap node internal memiliki satu atau - lebih cabang.
- Branches: Mewakili hasil dari keputusan atau pengujian, mengarah ke node lain.
- Leaf Nodes: Mewakili keputusan atau prediksi akhir. Tidak ada pembagian lebih lanjut pada node ini.

#### Bagimana Decision Tree bekerja?
Proses pembuatan decision tree melibatkan:

- Memilih Atribut Terbaik: Menggunakan metrik seperti Gini impurity, entropy, atau information gain, atribut terbaik untuk membagi data dipilih.
- Membagi Dataset: Dataset dibagi menjadi subset berdasarkan atribut yang dipilih.
- Mengulangi Proses: Proses diulangi secara rekursif untuk setiap subset, menciptakan node internal atau leaf node baru hingga kriteria penghentian terpenuhi (misalnya, semua instance dalam satu node termasuk dalam kelas yang sama atau kedalaman yang telah ditentukan tercapai).

#### Metrics untuk membagi
- Gini Impurity: Mengukur kemungkinan klasifikasi yang salah dari instance baru jika diklasifikasikan secara acak menurut distribusi kelas dalam dataset.
- Entropy: Mengukur jumlah ketidakpastian atau impurity dalam dataset
- Information Gain: Mengukur pengurangan entropy atau Gini impurity setelah dataset dibagi berdasarkan atribut.

#### Apa itu SVM (Support Vector Machine)?
Support Vector Machine (SVM) adalah algoritma pembelajaran mesin yang digunakan untuk klasifikasi dan regresi. SVM bekerja dengan menemukan hyperplane di ruang dimensi tinggi yang secara jelas memisahkan titik data dari kelas yang berbeda. Tujuan utama dari SVM adalah untuk menemukan hyperplane yang memiliki margin terbesar (jarak terjauh) antara titik data dari kedua kelas.

#### Bagaimana SVM Bekerja?

- Hyperplane: Dalam konteks SVM, hyperplane adalah garis atau permukaan yang memisahkan data menjadi kelas yang berbeda. Dalam dua dimensi, ini adalah garis; dalam tiga dimensi, ini adalah bidang.
- Support Vectors: Titik data yang paling dekat dengan hyperplane disebut support vectors. Mereka adalah elemen penting dari dataset karena mereka menentukan posisi dan orientasi hyperplane.
- Margin: Jarak antara hyperplane dan support vectors disebut margin. SVM mencoba memaksimalkan margin ini untuk memastikan bahwa hyperplane memisahkan kelas dengan baik.

#### Jenis-Jenis Kernel dalam SVM

Kernel adalah fungsi yang mengubah data ke dalam bentuk yang lebih tinggi dari dimensi yang membuatnya lebih mudah untuk memisahkan dengan hyperplane. Beberapa jenis kernel yang umum digunakan adalah:

- Linear Kernel: Digunakan saat data dapat dipisahkan secara linear. Fungsi kernel linear adalah:
    $K(x_i, x_j) = x_i \cdot x_j$
- Polynomial Kernel: Digunakan saat data tidak dapat dipisahkan secara linear. Fungsi kernel polynomial adalah:
    $K(x_i, x_j) = (\gamma x_i \cdot x_j + r)^d$
- Radial Basis Function (RBF): Digunakan saat data tidak dapat dipisahkan secara linear. Fungsi kernel RBF adalah:
    $K(x_i, x_j) = \exp(-\gamma ||x_i - x_j||^2)$
    
- Sigmoid Kernel: Digunakan saat data tidak dapat dipisahkan secara linear. Fungsi kernel sigmoid adalah:
    $K(x_i, x_j) = \tanh(\gamma x_i \cdot x_j + r)$

#### Kelebihan SVM

- Efektif dalam Dimensi Tinggi: SVM tetap efektif dalam ruang berdimensi tinggi dan bahkan dalam kasus di mana jumlah dimensi lebih besar dari jumlah sampel.
- Memori Efisien: SVM menggunakan subset dari titik pelatihan (support vectors) dalam proses keputusan, yang membuatnya memori efisien.
- Flexibilitas dengan Kernel: SVM dapat menggunakan berbagai fungsi kernel untuk memodelkan hubungan non-linear antara fitur.

#### Kekurangan SVM

- Waktu Pelatihan yang Lama: SVM bisa lambat dalam pelatihan, terutama untuk dataset yang sangat besar.
- Pemilihan Kernel dan Parameter: Memilih kernel yang tepat dan mengatur parameter SVM (seperti C dan ð›¾) dapat menjadi sulit dan memerlukan cross-validation.
- Kurang Efektif pada Dataset Besar dengan Noise: SVM kurang efektif jika terdapat banyak noise dan outliers dalam data karena mereka dapat mempengaruhi posisi hyperplane.

```python
from sklearn.ensemble import BaggingClassifier
def Bagging(method,X_train, y_train, X_test, y_test):
    le = LabelEncoder()
    y_train = le.fit_transform(y_train)
    y_test = le.fit_transform(y_test)
    acc = []
    index = []
    for n in range(3, 50, 2):
        clf = BaggingClassifier(estimator=method, n_estimators=n, random_state=None)
        clf.fit(X_train, y_train.ravel())
        y_pred = clf.predict(X_test)
        accuracy = accuracy_score(y_pred, y_test)
        acc.append(accuracy)
        index.append(str(clf).replace('(', '').replace(')', ''))

    df = pd.DataFrame({
        'Model': index,
        'Accuracy': acc

    })
    return df

Bagging_naive = Bagging(GaussianNB(),X_train_std, y_train_std, X_test_std, y_test_std)
Bagging_naive
```

## RANDOM FOREST

#### Apa itu Random Forest?
Random Forest adalah algoritma pembelajaran ensemble yang menggunakan beberapa decision tree untuk meningkatkan performa prediksi dan mengurangi overfitting. Ini adalah salah satu metode paling populer dan kuat dalam pembelajaran mesin, baik untuk tugas klasifikasi maupun regresi.

#### Bagaimana Random Forest Bekerja?

1) Pengambilan Sampel Acak (Bootstrapping): Algoritma memulai dengan pengambilan sampel acak dengan penggantian (bootstrapping) dari dataset pelatihan asli untuk membuat beberapa subset data.
2) Membangun Decision Tree: Untuk setiap subset data, sebuah decision tree dibangun. Saat membangun setiap tree, hanya subset acak dari fitur yang dipertimbangkan untuk pembagian di setiap node (feature randomness). Ini membantu dalam menciptakan variasi di antara tree yang berbeda.
3) Prediksi Ensemble: Setelah semua decision tree dibangun, prediksi dilakukan dengan menggabungkan hasil dari semua tree. Untuk klasifikasi, prediksi akhir adalah kelas yang dipilih oleh mayoritas tree (voting mayoritas). Untuk regresi, prediksi akhir adalah rata-rata dari semua prediksi tree.

#### Keuntungan Random Forest

- Mengurangi Overfitting: Dengan menggabungkan hasil dari banyak decision tree, Random Forest mengurangi risiko overfitting yang sering terjadi pada single decision tree.
- Robustness: Tahan terhadap outlier dan data yang hilang.
- Flexibilitas: Dapat digunakan untuk tugas klasifikasi dan regresi.
- Feature Importance: Dapat memberikan informasi tentang pentingnya fitur (feature importance), membantu dalam memilih fitur yang relevan.

#### Kekurangan Random Forest
- Kompleksitas dan Waktu Komputasi: Membutuhkan lebih banyak sumber daya komputasi dan waktu untuk melatih dan membuat prediksi dibandingkan dengan decision tree tunggal.
- Kurang Interpretabilitas: Meskipun lebih akurat, hasil dari Random Forest lebih sulit untuk diinterpretasikan dibandingkan dengan single decision tree.

#### Langkah-langkah dalam Random Forest:

1) Membangun Forest:
    - Mengambil sampel acak dengan penggantian dari dataset asli untuk membuat beberapa subset.
    - Membangun decision tree untuk setiap subset dengan mempertimbangkan subset acak dari fitur untuk pembagian di setiap node.
2) Prediksi:
    - Untuk klasifikasi: Voting mayoritas dari prediksi semua tree.
    - Untuk regresi: Rata-rata dari semua prediksi tree.

```python
from sklearn.ensemble import RandomForestClassifier
def RandomForest(X_train, y_train, X_test, y_test):
    le = LabelEncoder()
    y_train = le.fit_transform(y_train)
    y_test = le.fit_transform(y_test)
    acc = []
    index = []
    for n in range(2, 20, 1):
        clf = RandomForestClassifier(n_estimators=10, max_depth=n, random_state=42)
        clf.fit(X_train, y_train.ravel())
        y_pred = clf.predict(X_test)
        accuracy = accuracy_score(y_pred, y_test)
        acc.append(accuracy)
        index.append(str(clf).replace('(', '').replace(')', ''))

    df = pd.DataFrame({
        'Model': index,
        'Accuracy': acc

    })
    return df

Random_forest = RandomForest(X_train_std, y_train_std, X_test_std, y_test_std)
Random_forest
```

&nbsp;&nbsp;&nbsp;&nbsp;Stacking melibatkan dua atau lebih *base learner* atau model dasar sebagai model level-1, dan meta-learner yang menggabungkan prediksi base learner disebut sebagai level-2. Model yang berbeda digunakan oleh base learner untuk belajar dari suatu dataset. Output dari masing-masing model base learner dikumpulkan untuk membuat dataset baru. Di dalam dataset yang baru, setiap sampel berhubungan dengan nilai pada dataset awal yang seharusnya diprediksi. Selanjutnya dataset tersebut digunakan oleh model meta-learner hingga mendapatkan hasil akhir.
